```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# generate the vector of probabilities 
probability = np.empty(6)
probability.fill(1/6)
x = np.arange(1, 7)

fig, ax = plt.subplots()
ax.scatter(x, probability)
ax.set_xlabel('outcomes')
ax.set_ylabel('probability')
ax.set_title('Probability Distribution', fontweight ="bold", fontsize=16)
```

```python
# generate the vector of cumulative probabilities 
cum_probability = np.cumsum(probability)
fig, ax = plt.subplots()
ax.scatter(x, cum_probability)
ax.set_xlabel('outcomes')
ax.set_ylabel('cum_probability')
ax.set_title('Cumulative Probability Distribution', fontweight ="bold", fontsize=16)
```

```python
# get the probability of the binomial distribution given the parameters 
# k (how many head), n (number of tosses), probability (0.5 for head)
#  https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html
# http://web.stanford.edu/class/archive/cs/cs109/cs109.1192/handouts/pythonForProbability.html

# the probability of observing 5 successes
from scipy import stats 
X = stats.binom(10, 0.5) # Declare X to be a binomial random variable
print(X.pmf(5))          # P(X = 5)
print(X.cdf(4))          # P(X <= 4)
print(X.mean())          # E[X]
print(X.var())           # Var(X)
print(X.std())           # Std(X)
print(X.rvs())           # Get a random sample from X
print(X.rvs(10))         # Get 10 random samples form X
```

```python
# the probability of observing 4, 5, 6, 7 successes
sum_p = 0
for k in np.arange(4, 8):
    sum_p += stats.binom.pmf(k, 10, 0.5)
print(sum_p)
```

```python
# another way
a = X.cdf(7) # P(X <= 7)
b = X.cdf(3) # P(X <= 3)

# the probability of observing 4, 5, 6, 7 successes
print(a - b)
```

```python
# visualize the probability distribution function of k

p_list = []
for k in np.arange(0, 11):
    p_list.append(stats.binom.pmf(k, 10, 0.5))

x = np.arange(0, 11)

fig, ax = plt.subplots()
ax.scatter(x, p_list)
ax.set_xlabel('k')
ax.set_ylabel('probability')
ax.set_title('Probability Distribution Function', fontweight ="bold", fontsize=16)
```

```python
# visualize the cumulative distribution function of k

p_list = []
for k in np.arange(0, 11):
    p_list.append(stats.binom.cdf(k, 10, 0.5))

x = np.arange(0, 11)

fig, ax = plt.subplots()
ax.scatter(x, p_list)
ax.set_xlabel('k')
ax.set_ylabel('probability')
ax.set_title('Cumulative Distribution Function', fontweight ="bold", fontsize=16)
```

## Expected Value
The expected value of a random variable is, loosely, the long-run average value of its outcomes when the number of repeated trials is large. For a discrete random variable, the expected value is computed as a weighted average of its possible outcomes whereby the weights are the related probabilities.

$E(Y) = y_1 p_1 + y_2 p_2 + \cdots + y_k p_k = \sum_{i=1}^k y_i p_i$
